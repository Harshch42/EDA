## Ratings

```python
ratings = df_2.groupby(['Aggregate rating' ,'Rating color', 'Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

### Observations :-
- When Rating is between 4.5 - 4.9 ---> Excellent
- When Rating is between 4.0 - 4.4 ---> Very Good
- When Rating is between 3.5 - 3.9 ---> Good
- When Rating is between 2.5 - 3.4 ---> Average
- When Rating is between 0.1 - 2.4 ---> Poor
- More than 2000 customers did not rate.

```python
sns.barplot(x = 'Aggregate rating' , y = 'Rating Count' , data = ratings)
plt.show()
```

```python
sns.barplot(x = 'Aggregate rating' , y = 'Rating Count' ,hue = 'Rating color' ,data = ratings , palette = ['white' , 'red' , 'orange' , 'yellow' , 'green' , 'green' ] )
plt.show()
```

### Observations :-
- Maximum Ratings are between 2.5 and 3.4
- Not Rated Count is very high

```python
sns.countplot(x = 'Rating color' ,data = ratings , palette = ['white' , 'red' , 'orange' , 'yellow' , 'green' , 'green' ] )
plt.show()
```

```python
df_2[df_2['Aggregate rating'] == 0].groupby('Country').size().reset_index().rename(columns = {0 : 'Rating Count'})
```

### Observations :-

### Top 10 Cuisines

```python
cuisines = df[['Cuisines']].groupby('Cuisines').size().reset_index().rename(columns = {0 : 'Count'})
```

```python
cuisines.head()
```

```python
top_cuisines = cuisines.sort_values(by = 'Count' , ascending=False).head(10).reset_index().drop(columns = ['index'])
```

```python
top_cuisines
```

### Observations :- 
- North Indian , Chinese, Fast Food and Mughlai are most Preferred Cuisines.

## Cost

```python
df_2.columns
```

```python
top_cuisines
```

```python
 cuisines_price = pd.merge(top_cuisines , df_2  , on = 'Cuisines' , how = 'inner')[['Cuisines' ,'Average Cost for two' ]]
```

```python
cuisines_price
```

```python
cuisines_average_price = cuisines_price.groupby('Cuisines')['Average Cost for two'].agg(Average_Cost='mean').reset_index()
```

```python
cuisines_average_price = cuisines_average_price.sort_values(by = 'Average_Cost' , ascending=False)
```

```python
sns.barplot(y='Cuisines', x='Average_Cost', data=cuisines_average_price)
plt.xlabel('Average Cost')
plt.ylabel('Cuisines')
plt.title('Average Cost of Cuisines')
plt.show()
```

### Observations :-
- Average Cost of North Indian , Mughlai and Chinese food  in very high.
- Average Cost of Street food and Fast food is low.
